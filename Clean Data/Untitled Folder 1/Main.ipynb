```python
# Importing modules

import pandas as pd
import emoji
import re
import spacy
from spacy.matcher import Matcher
```

```python
df = pd.read_json('sample1.json')
df[1] = df[0] # copy 
df.set_axis(['Original', 'Edited'], axis=1, inplace=True)
df.head()
```

```python
column = 'Edited'

for i in df[column].index:
    text = ''.join(' ' + char + ' ' if (char in emoji.EMOJI_DATA or char == '\n') else char for char in df[column][i]) # Add spacing in front and back of emoji
    text = text.splitlines(True) # Split by '\n'. Return list of strings.
    text = [string.partition('\n') for string in text] # Partion by '\n'. Return list of sublists of strings.
    text = [string for sublist in text for string in sublist] # Remove inner list
    text = [string.strip(' ') for string in text] # Remove leading and trailing whitespaces.
    text = [re.sub(' +', ' ', string) for string in text] # Remove double spacing
    text = [string for string in text if string] # Remove strings with '' (nothing)
    df[column][i] = text
    print(df[column][i])
```

```python
# Emoji counter

column = 'Original'

emoji_list = []

for i in df[column].index:
    for char in df[column][i]:
        if char in emoji.EMOJI_DATA:
            emoji_list.append(char)

df_emoji = pd.DataFrame({'Emoji': emoji_list})
df_emoji = df_emoji.Emoji.value_counts().reset_index().rename(columns={'index' : 'Emoji', 'Emoji' : 'Count'})

for i in df_emoji.index:
    print(df_emoji['Emoji'][i], ' ', df_emoji['Count'][i],)
```

```python
# Here, we are going to be doing Natural Language Processing (NLP)

# Preprocessing text

df_text = df['Edited'].copy()
for i in df_text.index:
    for text in df_text[i]:
        if 'kiasu' in text:
            df_text[i].remove(text)
    df_text[i] = ' '.join(df_text[i])
    df_text[i] = re.sub('\n', '', df_text[i])
    df_text[i] = re.sub(' +', ' ', df_text[i])
    df_text[i] = df_text[i].strip()
    print(df_text[i])
```

```python
# Load pipeline
nlp = spacy.load("en_core_web_md")

doc = nlp(df_text[0])

for token in doc:
    print(token.text, token.pos_, token.dep_, token.head.text)
```

```python
for ent in doc.ents:
    print (ent.text, ent.label_)
```
